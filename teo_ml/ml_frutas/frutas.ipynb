```python
import pandas as pd
from sklearn import tree
im0port matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('dados_frutas.xlsx')
```

```python
df.head()
```

```python
arvore = tree.DecisionTreeClassifier(random_state=42)
```

```python
y = df['Fruta']
```

```python
caracteristicas = ['Arredondada', 'Suculenta', 'Vermelha', 'Doce']
```

```python
X = df[caracteristicas]
```

```python
arvore.fit(X, y)
```

```python
arvore.predict([[1,1,0,0]])
```

```python
tree.plot_tree(arvore, feature_names=caracteristicas, class_names=arvore.classes_, filled=True)
```

```python
arvore.predict_proba([[0,0,0,0]])
```
