```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt 
```

```python
df = pd.read_parquet('dados_clones.parquet')
```

```python
df
```

```python
df.columns
```

```python
features = ['Massa(em kilos)',
       'Estatura(cm)']
```

```python
target = ['Status ']
```

```python
X = df[features]
```

```python
y = df[target]
```

```python
model = tree.DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X, y)
```

```python
tree.plot_tree(model, feature_names=features, class_names=model.classes_, filled=True, max_depth=3)
```

É possível fazer a análise de que não é o general que está trazendo problemas e sim alguns lotes de clones defeituosos
