```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('dados_cerveja.xlsx')
```

```python
df.head()
```

```python
features = ['temperatura', 'copo','espuma','cor']
```

```python
X = df[features]
```

```python
target = ['classe']
```

```python
y = df[target]
```

```python
model = tree.DecisionTreeClassifier()
```

```python
X = X.replace({
    'mud': 1,'pint': 2, 'sim': 1, 'não': 0, 'escura': 1, 'clara': 0,
})
```

```python
model.fit(X, y)
```

```python
tree.plot_tree(model, feature_names=features, class_names=model.classes_, filled=True)
```
